```python

x = np.linspace(0, 10, 100)   # tạo 100 điểm từ 0 -> 10
y = np.sin(x)                 # tính sin(x)

plt.plot(x, y)                # vẽ đồ thị sin(x)
plt.title("Đồ thị hàm sin")
plt.xlabel("x")
plt.ylabel("sin(x)")
plt.show()
```

```python
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
plt.plot(x, y1, label = "Ham sin")
plt.plot(x, y2, label = "Ham cos")
plt.legend()
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,5))   # khung vẽ to hơn
# Ô 1 (hàng 1, cột 2, vị trí 1)
plt.subplot(2,1,1)
plt.plot([1,2,3],[1,4,9])
plt.title("Subplot 1")
# Ô 2 (hàng 1, cột 2, vị trí 2)
plt.subplot(2,2,1)
plt.plot([1,2,3],[1,2,3])
plt.title("Subplot 2")
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(8,6))
# Subplot 1
plt.subplot(2,1,1)   # 2 hàng, 1 cột, vị trí 1
plt.plot([1,2,3],[1,4,9], color = "m")
plt.title("Biểu đồ 1")
plt.xlabel("X")
plt.ylabel("Y")
# Subplot 2
plt.subplot(2,1,2)   # 2 hàng, 1 cột, vị trí 2
plt.plot([1,2,3],[1,2,3], color = "k")
plt.title("Biểu đồ 2")
plt.xlabel("X")
plt.ylabel("Y")
# Điều chỉnh layout
plt.tight_layout()
plt.show()
```

```python
x = [1, 2, 3, 4, 5]
y1 = [1, 2, 3, 4, 5]
y2 = [2, 3, 4, 5, 6]
y3 = [3, 4, 5, 6, 7]
y4 = [4, 5, 6, 7, 8]
plt.plot(x, y1, linestyle="-", label="Solid (-)")
plt.plot(x, y2, linestyle="--", label="Dashed (--)")
plt.plot(x, y3, linestyle="-.", label="Dash-dot (-.)")
plt.plot(x, y4, linestyle=":", label="Dotted (:)")
plt.grid(True)
plt.title("Ví dụ các kiểu linestyle")
plt.legend()
plt.show()

```

```python
x = np.linspace(0, 2*np.pi, 20)
y = np.sin(x)

plt.plot(x, y, color='m', linestyle='--', linewidth=2,
         marker='s', markersize=8, alpha=1, label='sin(x)')

plt.title("Ví dụ hàm plot() trong Python")
plt.xlabel("x (rad)")
plt.ylabel("Giá trị")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np 
import matplotlib.pyplot as plt  

# Compute the x and y coordinates for points on a sine curve 
x = np.arange(0, 3 * np.pi, 0.1) 
y = np.sin(x) 
plt.title("sine wave form") 

# Plot the points using matplotlib 
plt.plot(x, y, color='k', linestyle='--') 
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Tạo mảng x từ 0 đến 3π với bước 0.1
x = np.arange(0, 3 * np.pi, 0.1)

# Tính sin(x) và cos(x)
y_sin = np.sin(x)
y_cos = np.cos(x)

# Vẽ đồ thị sin(x)
plt.plot(x, y_sin, label="sin(x)", color="b")

# Vẽ đồ thị cos(x)
plt.plot(x, y_cos, label="cos(x)", color="r")

# Thêm tiêu đề, trục, chú thích
plt.title("Đồ thị hàm sin(x) và cos(x)")
plt.xlabel("x (rad)")
plt.ylabel("Giá trị")
plt.grid(True)
plt.legend()
plt.show()
```

```python
x = ["Mon", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun"]
y = [10, 12, 8, 15, 20, 18, 25]

# Tạo figure với 2 biểu đồ cạnh nhau
plt.figure(figsize=(10,4))

# Line chart
plt.subplot(1,2,1)
plt.plot(x, y, marker="o", color="b")
plt.title("Line Chart")
plt.xlabel("Day")
plt.ylabel("Value")
# Bar chart
plt.subplot(1,2,2)
plt.bar(x, y, color="orange")
plt.title("Bar Chart")
plt.xlabel("Day")
plt.ylabel("Value")
# Hiển thị
plt.tight_layout()
plt.show()
```
