```python
x = [1, 2, 3, 4, 5]
y1 = [1, 2, 3, 4, 5]
y2 = [2, 3, 4, 5, 6]
y3 = [3, 4, 5, 6, 7]
y4 = [4, 5, 6, 7, 8]
plt.plot(x, y1, linestyle="-", label="Solid (-)")
plt.plot(x, y2, linestyle="--", label="Dashed (--)")
plt.plot(x, y3, linestyle="-.", label="Dash-dot (-.)")
plt.plot(x, y4, linestyle=":", label="Dotted (:)")
plt.grid(True)
plt.title("V d cc kiu linestyle")
plt.legend()
plt.show()

```

```python
x = np.linspace(0, 2*np.pi, 20)
y = np.sin(x)

plt.plot(x, y, color='m', linestyle='--', linewidth=2,
         marker='s', markersize=8, alpha=1, label='sin(x)')

plt.title("V d hm plot() trong Python")
plt.xlabel("x (rad)")
plt.ylabel("Gi tr")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np 
import matplotlib.pyplot as plt  

# Compute the x and y coordinates for points on a sine curve 
x = np.arange(0, 3 * np.pi, 0.1) 
y = np.sin(x) 
plt.title("sine wave form") 

# Plot the points using matplotlib 
plt.plot(x, y, color='k', linestyle='--') 
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# To mng x t 0 n 3 vi bc 0.1
x = np.arange(0, 3 * np.pi, 0.1)

# Tnh sin(x) v cos(x)
y_sin = np.sin(x)
y_cos = np.cos(x)

# V  th sin(x)
plt.plot(x, y_sin, label="sin(x)", color="b")

# V  th cos(x)
plt.plot(x, y_cos, label="cos(x)", color="r")

# Thm tiu , trc, ch thch
plt.title(" th hm sin(x) v cos(x)")
plt.xlabel("x (rad)")
plt.ylabel("Gi tr")
plt.grid(True)
plt.legend()
plt.show()
```

```python
x = ["Mon", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun"]
y = [10, 12, 8, 15, 20, 18, 25]

# To figure vi 2 biu  cnh nhau
plt.figure(figsize=(10,4))

# Line chart
plt.subplot(1,2,1)
plt.plot(x, y, marker="o", color="b")
plt.title("Line Chart")
plt.xlabel("Day")
plt.ylabel("Value")
# Bar chart
plt.subplot(1,2,2)
plt.bar(x, y, color="orange")
plt.title("Bar Chart")
plt.xlabel("Day")
plt.ylabel("Value")
# Hin th
plt.tight_layout()
plt.show()
```
